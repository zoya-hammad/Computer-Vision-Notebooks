```python
# Our Setup, Import Libaries, Create our Imshow Function
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function
def imshow(title = "Image", image = None, size = 6):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
template = cv2.imread('waldo.png')
imshow('Template', template, 4)
```

```python
image = cv2.imread('wheres_waldo.webp')
imshow('Image',image, 15)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_temp = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
```

```python
result = cv2.matchTemplate(gray,gray_temp, cv2.TM_CCOEFF)
# Compares the template image with every possible location in the larger image.
# Produces a result heatmap where brighter spots indicate a better match.
```

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
```

cv2.minMaxLoc(result) returns:
- min_val, max_val: Minimum and maximum similarity scores.
- min_loc, max_loc: Locations of these values.

```python
top_left = max_loc

w, h = gray_temp.shape[::-1]
bottom_right = (top_left[0] + w, top_left[1] + h)

cv2.rectangle(image, top_left, bottom_right, (255,0,0), 5)
```

```python
imshow('Where is Waldo?', image, 20)
```
